# **Menggabungkan 2 data**

Merge digunakan untuk menggabungkan dua atau lebih DataFrame berdasarkan satu atau beberapa kunci atau kolom yang digunakan sebagai kunci penggabungan.

tujuan dari saya menggabungan 2 data studentAssesment dengan student Info adalah untuk mengetahui nilai akhir yang ada di student assesment  dari data ID student dan modul yang ada di student info


Kode dibawah ini adalah perintah yang digunakan untuk mengimpor modul pyplot dari perpustakaan Matplotlib ke dalam program Python Anda. Matplotlib adalah salah satu perpustakaan yang paling populer untuk membuat grafik dan visualisasi data dalam Python.

```python
import matplotlib.pyplot as plt
```

kode dibawah ini untuk mengambil data yang sudah ada dalam variabel studentAssessment dan menggunakan fungsi pd.DataFrame() dari perpustakaan Pandas untuk mengubahnya menjadi objek DataFrame yang baru dengan nama studentAssessment. Dengan ini, kita dapat menggunakan berbagai fungsi Pandas untuk mengakses dan memanipulasi data yang ada dalam studentAssessment.

```python
studentAssessment = pd.DataFrame(studentAssessment)
studentInfo = pd.DataFrame(studentInfo)
```

kode dibawah ini digunakan untuk melakukan operasi merge antara dua objek DataFrame, yaitu studentAssessment dan studentInfo, berdasarkan kolom 'id_student' dengan tipe merge 'inner'. Ini adalah salah satu cara yang umum digunakan dalam Pandas untuk menggabungkan dua DataFrames berdasarkan kriteria tertentu.

```python
# Melakukan merge atau penggabungan 2 data
merged_data = pd.merge(studentAssessment, studentInfo, on='id_student', how='inner')
```

Kode yang dibawah ini untuk membuat dan menampilkan grafik batang (bar chart) berdasarkan data yang telah digabungkan dalam objek DataFrame merged_data

```python
# Membuat grafik batang (bar chart)
plt.figure(figsize=(10, 6))

# Mengelompokkan data berdasarkan code_module dan final_result
grouped_data = merged_data.groupby(['code_module', 'final_result']).size().unstack(fill_value=0)

# Membuat grafik batang
grouped_data.plot(kind='bar', stacked=True)

# Menambahkan label dan judul
plt.xlabel('code_module')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Grafik Hasil Final Berdasarkan Code Module')

# Menampilkan grafik
plt.legend(title='Final Result')
plt.show()
```

```python
#Kode ini menghitung jumlah kemunculan setiap kategori dalam kolom 'final_result' dalam DataFrame merged_data.
#Hasilnya adalah objek Series yang berisi jumlah siswa untuk setiap hasil akhir
#(misalnya, 'Distinction', 'Pass', 'Fail', dsb.).

final_result_counts = merged_data['final_result'].value_counts()

# Mengatur ukuran gambar
plt.figure(figsize=(7, 4))

# Membuat grafik batang
plt.bar(final_result_counts.index, final_result_counts, color=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])

# Memberikan judul dan label pada grafik
plt.title('Distribusi Hasil Akhir Siswa')
plt.xlabel('Hasil Akhir Siswa')
plt.ylabel('Jumlah Siswa')

# Menampilkan grafik
plt.xticks(rotation=45)  # Untuk memutar label sumbu x agar lebih mudah dibaca
plt.tight_layout()  # Untuk mengatur tata letak grafik agar rapi
plt.show()
```
